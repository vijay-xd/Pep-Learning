```python
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt

# Pick the best available device. This machine has no CUDA GPU, so we fall
# back to CPU. float16 is not well supported on CPU, so use float32 there.
device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.float16 if device == "cuda" else torch.float32
print(f"Using device={device}, dtype={dtype}")

# Load the Stable Diffusion model from Hugging Face
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=dtype,
    safety_checker=None,
).to(device)

# Prompt for generation
prompt = "a futuristic city skyline at sunset, digital art"

# Generate image. autocast only helps (and is only supported) on CUDA;
# on CPU we just run the pipeline directly.
# NOTE: CPU inference is slow - reduce steps for a faster test run.
if device == "cuda":
    with torch.autocast("cuda"):
        image = pipe(prompt, num_inference_steps=25).images[0]
else:
    image = pipe(prompt, num_inference_steps=25).images[0]

# Save the image to a file
image.save("output_image.png")

# Display the image
plt.imshow(image)
plt.axis("off")
plt.title("Generated by Stable Diffusion")
plt.show()
```

```python
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt

# Prompt input from user (falls back to a default if run non-interactively,
# e.g. via `jupyter nbconvert --execute`, where stdin isn't available)
try:
    prompt = input("enter the prompt here: ") or "a futuristic city skyline at sunset, digital art"
except EOFError:
    prompt = "a futuristic city skyline at sunset, digital art"
    print(f"No interactive input available, using default prompt: {prompt!r}")

# Pick the best available device (this machine has no CUDA GPU -> CPU/float32)
device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.float16 if device == "cuda" else torch.float32

# Load the pretrained Stable Diffusion model
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=dtype,
    safety_checker=None,
).to(device)

# Generate the image using the user's prompt
if device == "cuda":
    with torch.autocast("cuda"):
        image = pipe(prompt, num_inference_steps=25).images[0]
else:
    image = pipe(prompt, num_inference_steps=25).images[0]

# Save the image to a file
image.save("generated_image.png")
print("Image saved as 'generated_image.png'.")

# Display the image
plt.imshow(image)
plt.axis("off")
plt.title(f"Generated for prompt:\n{prompt}")
plt.show()
```

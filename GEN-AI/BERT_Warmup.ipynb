Tokenize a sentence

```python
#Convert Tokens into IDs
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "AI is amazing"

inputs = tokenizer(text)

print(inputs["input_ids"])
```

```python
#Convert IDs Back to Tokens
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

ids = [101, 9932, 2003, 6429, 102]

tokens = tokenizer.convert_ids_to_tokens(ids)

print(tokens)
```

```python
#Load BERT Model
from transformers import AutoModel

model = AutoModel.from_pretrained("bert-base-uncased")

print("BERT Loaded")
```

```python
#Generate Embedding
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

text = "AI is amazing"

inputs = tokenizer(text, return_tensors="pt")

outputs = model(**inputs)

print(outputs.last_hidden_state.shape)
```

```python
#view embeddig in one token
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

text = "AI is amazing"

inputs = tokenizer(text, return_tensors="pt")

outputs = model(**inputs)

embedding = outputs.last_hidden_state[0][1]

print(embedding.shape)
```

```python
#see actual embeddigng numbers
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

text = "AI"

inputs = tokenizer(text, return_tensors="pt")

outputs = model(**inputs)

embedding = outputs.last_hidden_state[0][1]

print(embedding[:10])
```

```python
#understanding attention mask
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "AI is amazing"

inputs = tokenizer(text)

print(inputs["attention_mask"])
```

```python
#Two-Sentence Input for NSP Understanding
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text1 = "What is AI?"
text2 = "AI means Artificial Intelligence."

inputs = tokenizer(text1, text2)

tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"])

print(tokens)
```

```python
#Extract [CLS] Embedding
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

text = "AI is amazing"

inputs = tokenizer(text, return_tensors="pt")

outputs = model(**inputs)

cls_embedding = outputs.last_hidden_state[:,0,:]

print(cls_embedding)
```

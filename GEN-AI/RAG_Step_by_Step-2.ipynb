## Step 2 — Chunking

````python
text = '''
AI is transforming education.
RAG helps reduce hallucination.
Embeddings convert text into vectors.
'''

chunks = text.split(".")

print(chunks)
````

## Step 3 — Remove Empty Chunks

````python
chunks = [chunk.strip() for chunk in chunks if chunk.strip()]

print(chunks)
````

## Step 5 — Understanding Embeddings

````python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

sentence = "AI is transforming education"

embedding = model.encode(sentence)

print(embedding.shape)
````

## Step 6 — Embedding Multiple Chunks

````python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

chunks = [
    "AI is transforming education",
    "RAG helps reduce hallucination",
    "Embeddings convert text into vectors"
]

embeddings = model.encode(chunks)

print(embeddings.shape)
````

## Step 7 — User Query

````python
query = "What reduces hallucination?"

print(query)
````

## Step 8 — Query Embedding

````python
query_embedding = model.encode(query)

print(query_embedding.shape)
print(query_embedding)
````


## Step 9 — Install Scikit-learn

## Step 10 — Similarity Search

````python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(
    [query_embedding],
    embeddings
)

print(similarities)
````

## Step 11 — Retrieve Best Matching Chunk

````python
best_match_index = similarities.argmax()

print(chunks[best_match_index])
````

## Step 12 — Prompt Augmentation

````python
retrieved_chunk = chunks[best_match_index]

prompt = f'''
Context:
{retrieved_chunk}

Question:
{query}
'''

print(prompt)
````

## Step 13 — Simulated LLM Response

````python
answer = f'''
Based on the context,
RAG helps reduce hallucination.
'''

print(answer)
````

## Step 14 — Complete Mini RAG Pipeline

````python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Document chunks
chunks = [
    "AI is transforming education",
    "RAG helps reduce hallucination",
    "Embeddings convert text into vectors"
]

# Load embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Create embeddings
chunk_embeddings = model.encode(chunks)

# User query
query = "How to reduce hallucination?"

# Query embedding
query_embedding = model.encode(query)

# Similarity search
similarities = cosine_similarity(
    [query_embedding],
    chunk_embeddings
)

# Retrieve best chunk
best_match = similarities.argmax()

retrieved_chunk = chunks[best_match]

# Augmented prompt
prompt = f'''
Context:
{retrieved_chunk}

Question:
{query}
'''

print(prompt)
````
